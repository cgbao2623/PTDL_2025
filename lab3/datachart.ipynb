```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_csv('pima-indians-diabetes.data.csv', header=None)
df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
              'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

# 1. Phân phối Outcome
plt.figure(figsize=(6,4))
sns.countplot(x=df['Outcome'])
plt.title('Phân phối biến Outcome')
plt.show()

# 2. Histogram cho các đặc trưng
df.hist(bins=20, figsize=(12,10))
plt.suptitle('Histogram của các đặc trưng')
plt.show()

# 3. Boxplot theo Outcome
features = df.columns[:-1]
plt.figure(figsize=(15,10))
for i, feature in enumerate(features, 1):
    plt.subplot(3,3,i)
    sns.boxplot(x='Outcome', y=feature, data=df)
    plt.title(f'Boxplot của {feature} theo Outcome')
plt.tight_layout()
plt.show()

# 4. Heatmap tương quan
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Ma trận tương quan')
plt.show()

# 5. Scatter plot: Glucose vs BMI
plt.figure(figsize=(8,6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df, alpha=0.7)
plt.title('Glucose vs BMI (màu theo Outcome)')
plt.show()

# 6. Pairplot cho một số biến quan trọng
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome', diag_kind='kde')
plt.suptitle('Pairplot của các biến quan trọng', y=1.02)
plt.show()

# 7. Kiểm tra giá trị 0 bất thường (ví dụ: Glucose, BloodPressure không thể bằng 0)
zero_check = (df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] == 0).sum()
print("Số giá trị 0 không hợp lý trong các cột:\n", zero_check)
```
