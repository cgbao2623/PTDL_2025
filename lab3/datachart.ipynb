```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# c d liu
df = pd.read_csv('pima-indians-diabetes.data.csv', header=None)
df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
              'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

# 1. Phn phi Outcome
plt.figure(figsize=(6,4))
sns.countplot(x=df['Outcome'])
plt.title('Phn phi bin Outcome')
plt.show()

# 2. Histogram cho cc c trng
df.hist(bins=20, figsize=(12,10))
plt.suptitle('Histogram ca cc c trng')
plt.show()

# 3. Boxplot theo Outcome
features = df.columns[:-1]
plt.figure(figsize=(15,10))
for i, feature in enumerate(features, 1):
    plt.subplot(3,3,i)
    sns.boxplot(x='Outcome', y=feature, data=df)
    plt.title(f'Boxplot ca {feature} theo Outcome')
plt.tight_layout()
plt.show()

# 4. Heatmap tng quan
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Ma trn tng quan')
plt.show()

# 5. Scatter plot: Glucose vs BMI
plt.figure(figsize=(8,6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df, alpha=0.7)
plt.title('Glucose vs BMI (mu theo Outcome)')
plt.show()

# 6. Pairplot cho mt s bin quan trng
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome', diag_kind='kde')
plt.suptitle('Pairplot ca cc bin quan trng', y=1.02)
plt.show()

# 7. Kim tra gi tr 0 bt thng (v d: Glucose, BloodPressure khng th bng 0)
zero_check = (df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] == 0).sum()
print("S gi tr 0 khng hp l trong cc ct:\n", zero_check)
```
